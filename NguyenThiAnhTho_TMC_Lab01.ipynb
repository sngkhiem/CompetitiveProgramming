```python
import math
import numpy
import matplotlib.pyplot as p

real_value = numpy.exp(-5)
l1 = []
l2 = []


def first_way(x):
	res = 0;
	pre_res = 0
	e_t = 0
	e_a = 0
	print("Calculation using first way: ")
	print("%s%20s%15s%20s" % ("Interation","Approximation","E_a","E_t"))
	for n in range (0,20):
		pre_res = res;
		res += math.pow(-x,n)/math.factorial(n)
		l1.append(res)
		e_t = ((real_value-res)/real_value)
		e_a = ((res - pre_res)/res)
		print("%9d%20.9f%20.9f%20.9f" % (n,res,e_a,e_t))


def second_way(x):
    res = 0
    pre_res = 10
    e_t = 0
    e_a = 0
    print("Calculation using second way: ")
    print("%s%20s%15s%20s" % ("Interation","Approximation","E_a","E_t"))
    for n in range (0,20):
        pre_res = res
        if res != 0:
            flip_res = 1/res
        else:
            flip_res = 0
        flip_res += (math.pow(x,n)/math.factorial(n))
        res = 1/flip_res
        l2.append(res)
        e_t = ((real_value-res)/real_value)
        e_a = ((res - pre_res)/res)
        print("%9d%20.10f%20.10f%20.10f" % (n,res,e_a,e_t))

first_way(5)
print("\n")
p.subplot(2,1,1)
p.plot(l1)
p.title("The first_way_Code by Nguyen Thi Anh Tho_Lab01_TMC_Ex01")
second_way(5)
print("\n")
```

```python
p.subplot(2,1,2)
p.plot(l2)
p.title("The second_way_Code by Nguyen Thi Anh Tho_Lab01_TMC_Ex01")
```

```python
import math
import numpy
import matplotlib.pyplot as p

def seriesWay(x):
    s = 0
    for i in range(10):
        s += (numpy.power(x ,i*2) / math.factorial(i*2)) * numpy.power(-1, i)
    return s

var = numpy.pi * 0.3
print(math.cos(var))
print(seriesWay(var))

x = numpy.arange(-2*numpy.pi, 2*numpy.pi, 0.1)
y1 = seriesWay(x)
y2 = numpy.cos(x)
p.subplot(2, 1, 1)
p.plot(x, y1)
p.title("F(x)_Code by Nguyen Thi Anh Tho_Lab01_TMC_Ex01")
```

```python
p.subplot(2, 1, 2)
p.plot(x, y2)
p.title("Cos(x)_Code by Nguyen Thi Anh Tho_Lab01_TMC_Ex01")
p.show()
```

```python
from __future__ import division
import numpy as np
import math
import matplotlib.pyplot as p


def myFunc(x):
    return 1/np.power(x,2)

def myFunc1stDerivative(x):
    return -2 / np.power(x, 3)

def myFunc2ndDerivative(x):
    return 6 / np.power(x, 4)

def myFunc3rdDerivative(x):
    return -24 / np.power(x, 5)

def myFunc4ndDerivative(x):
    return 120 / np.power(x, 6)

def fourthOrderTaylorSeries(x, x0):
    return \
        myFunc(x0) + \
        (myFunc1stDerivative(x0) / math.factorial(1)) * np.power((x - x0), 1) + \
        (myFunc2ndDerivative(x0) / math.factorial(2)) * np.power((x - x0), 2) + \
        (myFunc3rdDerivative(x0) / math.factorial(3)) * np.power((x - x0), 3) + \
        (myFunc4ndDerivative(x0) / math.factorial(4)) * np.power((x - x0), 4)

x = np.arange(-30, 30, 0.1)
y1 = fourthOrderTaylorSeries(0, x)
p.subplot(2,1,1)
p.plot(x, y1)
p.title("F(x)_Code by Nguyen Thi Anh Tho_Lab01_TMC_Ex01")
```

```python
y2 = myFunc(x)
p.subplot(2,1,2)
p.plot(x, y2)
p.title("\nTrue value_Code by Nguyen Thi Anh Tho_Lab01_TMC_Ex01")
p.show()
```
